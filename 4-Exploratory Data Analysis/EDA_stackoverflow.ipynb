```python
fulltime_datascientist=df.loc[(df["DevType"] == "Data scientist or machine learning specialist") & (df["Employment"]=="Employed full-time")]
fulltime_datascientist




```

```python
def histo(a,x):
    plt.figure(figsize=(10,10))
    plt.hist(a["ConvertedSalary"])
    plt.title("Annual salaries of "+x)
    plt.xlabel("annual salary in $ ")
histo(fulltime_datascientist,"Data scientist or machine learning specialist")
```

```python
# The salaries of data scientists follow practically the same distribution as the previously viewed tech jobs salaries
```

```python
#Now lets see what variables act on your salary as a datascientist
def corr(var):
    by_var=fulltime_datascientist[[var,"ConvertedSalary"]].dropna()
    return (by_var.groupby([var]).mean().sort_values(by="ConvertedSalary",ascending=False))
corr("Country")
```
